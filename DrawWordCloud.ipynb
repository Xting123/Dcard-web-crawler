```python
import json
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from matplotlib.font_manager import FontProperties
```

```python
file ="JiebaData.json"
with open(file, 'r', encoding='utf-8') as obj:
    data = json.load(obj)
print(data)
```

```python
indata=data[0]
for i in range(len(indata)):
    mask = np.array(Image.open('cutecloud.jpg'))
    wordcloud = WordCloud(background_color="white",
                          width=500, height=350, 
                          margin=2, font_path="simhei.ttf", 
                          mask=mask).generate(str(list(indata.values())[i]))
    plt.figure(figsize=(5,5))
    plt.imshow(wordcloud)
    myfont = FontProperties(fname=r'./NotoSansCJK-Black.ttc')
    plt.title(str(list(indata.keys())[i]), fontproperties = myfont, fontsize = 18)
    plt.axis('off')
    plt.show()
#繪製成文字雲
```
