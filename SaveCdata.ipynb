```python
import json
import matplotlib.pyplot as plt
import pandas as pd
import jieba
import operator
import numpy as np
from matplotlib.font_manager import FontProperties
import requests
import re
import time
from bs4 import BeautifulSoup as bs
```

```python
file ="data.json"
with open(file, 'r', encoding='utf-8') as obj:
    data = json.load(obj)
```

```python
id=[]
for content in data:
    id.append(content['id'])
print(id)
```

```python
request_headers={
    'user-agent':""}

Cdata = {}
for i in range(len(id)):
    ID = str(id[i])
    url = 'https://www.dcard.tw/f/pet/p/'+ID
    reqs = requests.get(url)
    soup = bs(reqs.text,'lxml')
    title = soup.find('h1',{'class':'sc-1932jlp-0 cqaWIE'}).text #標題
    category = soup.find('article',{'class':'sc-1eorkjw-0 dqGaqV'}).text #內文
    Cdata[title] = category
    time.sleep(300)
print(Cdata)
```

```python
with open('CDATA.json', mode='w', encoding='utf-8')as file:
    json.dump(Cdata, file, ensure_ascii=False)
print("OK")
```
